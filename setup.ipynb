```python
import numpy as np
from data_utils import list_nii_paths, visualize_sample

import monai
from monai.losses import DiceLoss
from monai.metrics import DiceMetric

import torch
from torch import optim
from torch.utils.data import DataLoader, random_split
```

```python
from torch.utils.data import Dataset

class CancerNetPCa(Dataset):
    def __init__(self, img_path, mask_path, img_size=(256, 256), transform=None):
        self.img_path = img_path
        self.mask_path = mask_path
        self.img_size = img_size
        self.transform = transform

    def __len__(self):
        return len(self.img_path)

    def __getitem__(self, idx):
        img_path = self.img_path[idx]
        mask_path = self.mask_path[idx]

        img = nib.load(img_path).dataobj
        img = np.array(img).astype(np.uint8)

        mask = nib.load(mask_path).dataobj
        mask = np.array(mask).astype(np.uint8)

        slice_num = 9
        img_slice = img[:, :, slice_num]
        mask_slice = mask[:, :, slice_num]

        img_std = resize_image(normalize_image(img_slice), self.img_size)
        mask_std = resize_image(normalize_image(mask_slice), self.img_size)

        img_tensor = torch.from_numpy(img_std).float()
        mask_tensor = torch.from_numpy(mask_std).long()

        img_tensor = img_tensor.unsqueeze(0)
        mask_tensor = mask_tensor.unsqueeze(0)

        # apply transformations if provided (add later)
        #if self.transform:

        return img_tensor, mask_tensor
```

```python
import cv2

def resize_image(img, img_size=(256, 256)):
    """Resize images to desired size.
    Args:
        img: Image array.
        img_size: Desire image array resize.
    Yields:
        Resized image array.
    """
    return cv2.resize(img, img_size, interpolation=cv2.INTER_AREA)

def normalize_image(img):
    """Normalize images to [0, 1].
    Args:
        img: Image array.
    Yields:
        Normalized image array.
    """
    return img / img.max() if img.max() != 0 else img
```

```python
img_path = list_nii_paths('data/images')
mask_path = list_nii_paths('data/masks')

dataset = CancerNetPCa(img_path=img_path, mask_path=mask_path)
```

```python
import nibabel as nib

img_tensor, mask_tensor = dataset[0]

for i in range(3):
    img_tensor, mask_tensor = dataset[i]
    visualize_sample(img_tensor, mask_tensor, (7, 3))
```

## Training

```python
split = 0.7
# can implement shuffled indexing after if useful
num_items = len(dataset)
num_train = round(num_items * split)
num_val = round(num_items * (split / 4))
num_test = num_items - num_train - num_val
train_ds, val_ds, test_ds = random_split(dataset, [num_train, num_val, num_test])

train_loader = DataLoader(train_ds, batch_size=4, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=4)
test_loader = DataLoader(test_ds, batch_size=1)
```

```python
train_loader, test_loader, val_loader
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
from monai.networks.nets import SegResNet

segresnet = SegResNet(
    spatial_dims=2,
    blocks_down=[1, 2, 2, 4],
    blocks_up=[1, 1, 1],
    init_filters=16,
    in_channels=1,
    out_channels=1,
    dropout_prob=0.2,
)
```

```python
model = segresnet.to(device)

loss_function = DiceLoss(sigmoid=True)
optimizer = torch.optim.Adam(model.parameters(), 0.001)
max_epochs = 300
train_interval = 5
val_interval = 2

best_metric = 0
best_metric_epoch = 0
dice_metric = DiceMetric(include_background=True, reduction="mean")

early_stop = True
max_no_improvement = 25
no_improvement = 0

for epoch in range(max_epochs):
    model.train()
    epoch_loss = 0
    for batch_data in train_loader:
        inputs, labels = batch_data
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = loss_function(outputs, labels)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    if (epoch + 1) % train_interval == 0:
        print(f"epoch {epoch + 1} average loss: {epoch_loss:.4f}")

    if (epoch + 1) % val_interval == 0:
        model.eval()
        with torch.no_grad():
            val_dice = []
            for val_data in val_loader:
                val_inputs, val_labels = val_data
                val_inputs, val_labels = val_inputs.to(device), val_labels.to(device)
                val_outputs = model(val_inputs)
                val_outputs = torch.sigmoid(val_outputs)
                dice_metric(y_pred=val_outputs, y=val_labels)

            metric = dice_metric.aggregate().item()
            print(metric)
            val_dice.append(metric)
            dice_metric.reset()
            mean_val_dice = np.mean(val_dice)

            mean_val_dice = np.mean([score for score in val_dice if not np.isnan(score)])
            print(mean_val_dice)
            if mean_val_dice > best_metric:
                best_metric = mean_val_dice
                best_metric_epoch = epoch + 1
                torch.save(model.state_dict(), "best_segresnet2d.pth")
                no_improvement = 0
                print(f">>> Best dice: {best_metric} at epoch {best_metric_epoch}")
            else:
                no_improvement += 1
                if early_stop and no_improvement >= max_no_improvement:
                    print(f">>> Training stopped early at epoch {epoch+1}")
                    break
```

```python
model = segresnet.to(device)

model.load_state_dict(torch.load("best_segresnet2d.pth"))
model.eval()

with torch.no_grad():
    test_dice = []
    for test_data in test_loader:
        test_inputs, test_labels = test_data
        test_inputs, test_labels = test_inputs.to(device), test_labels.to(device)
        test_outputs = model(test_inputs)
        test_outputs = torch.sigmoid(test_outputs)
        visualize_sample(test_inputs.cpu(), test_labels.cpu(), (7, 3), pred_tensor=test_outputs.cpu())
        dice_metric(y_pred=test_outputs, y=test_labels)

    metric = dice_metric.aggregate().item()
    dice_metric.reset()

print(f"Test set dice score: {metric}")
```
